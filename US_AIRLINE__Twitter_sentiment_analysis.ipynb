```python
ech_tweet_train= []
for line in tweets_train:
    ech_tweet_train.append([wrds for wrds in re.split(' |,|!|\?', line)])
    #print(ech_tweet_train)
```

```python
ech_tweet_test = []
for line in tweets_test:
    ech_tweet_test.append([wrds for wrds in re.split(' |,|!|\?', line)])
    #print(ech_tweet_test)
ech_tweet_test
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet
def get_pos_tag(word):
    tag = pos_tag([word])[0][1]
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
```

```python
from nltk.corpus import stopwords
import string
# stops = set(stopwords.words('english'))
# punctuations = list(string.punctuation)
# stops.update(punctuations)
# stops, string.punctuation


punctuations = list(string.punctuation)
stop = stopwords.words('english')
stop += punctuations
```

```python
def clean_tweet(words):
    output_words = list()
    for w in words:
        if w.lower() not in stop and len(w) > 0:            
            clean_word = lemmatizer.lemmatize(w, pos = get_pos_tag(w))
            output_words.append(clean_word.lower())
    return output_words
```

```python
text_tweet_train = [(clean_tweet(text)) for text in ech_tweet_train]
text_tweet_train
```

```python
text_tweet_test = [(clean_tweet(text)) for text in ech_tweet_test]
text_tweet_test
```

```python
clean_text_tweet_train = [" ".join(text) for text  in text_tweet_train]
clean_text_tweet_test = [" ".join(text) for text  in text_tweet_test]
```

```python
print(clean_text_tweet_train)
print(clean_text_tweet_test)
```

```python
sentiment = [sentiment for sentiment  in train_data['airline_sentiment']]
```

```python
len(clean_text_tweet_train), len(sentiment), len(clean_text_tweet_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer(max_features = 2000, ngram_range=(1,3))
x_train_features = count_vec.fit_transform(clean_text_tweet_train, sentiment)
x_train_features = x_train_features.todense()
x_test_features = count_vec.transform(clean_text_tweet_test)
x_test_features = x_test_features.todense()
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(C=100, degree = 2, kernel = 'rbf')
svc.fit(x_train_features, sentiment)
```

```python
sentiment_pred = svc.predict(x_test_features)
sentiment_pred
```

```python
set(sentiment_pred)
```

```python
score = svc.score(x_test_features, sentiment_pred)
score
```
